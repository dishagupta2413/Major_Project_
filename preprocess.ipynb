```python
#to check stationarity
from statsmodels.tsa.stattools import adfuller

#function checks stationarity
def check_stationarity(timeseries):

    result=adfuller(timeseries)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])

    if result[1] <= 0.05:
        print('Stationary (Reject the null hypothesis)')
    else:
        print('Non-Stationary (Fail to reject the null hypothesis)')

check_stationarity(df_cleaned_closed)
```

```python
df_cleaned_closed_diff=df_cleaned_closed.diff().dropna()

check_stationarity(df_cleaned_closed_diff) #here d=1 as after one differencing it becomes stationary
```

```python

#calculation of values of p and q now
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

plot_acf(df_cleaned_closed_diff)
plot_pacf(df_cleaned_closed_diff)
plt.show()
```
